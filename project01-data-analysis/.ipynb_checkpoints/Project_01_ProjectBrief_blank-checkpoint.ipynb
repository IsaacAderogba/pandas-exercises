## Data Import and first Inspection

1. __Import__ the movies dataset from the CSV file "movies_complete.csv". __Inspect__ the data.

__Some additional information on Features/Columns__:

* **id:** The ID of the movie (clear/unique identifier).
* **title:** The Official Title of the movie.
* **tagline:** The tagline of the movie.
* **release_date:** Theatrical Release Date of the movie.
* **genres:** Genres associated with the movie.
* **belongs_to_collection:** Gives information on the movie series/franchise the particular film belongs to.
* **original_language:** The language in which the movie was originally shot in.
* **budget_musd:** The budget of the movie in million dollars.
* **revenue_musd:** The total revenue of the movie in million dollars.
* **production_companies:** Production companies involved with the making of the movie.
* **production_countries:** Countries where the movie was shot/produced in.
* **vote_count:** The number of votes by users, as counted by TMDB.
* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

### The Best and the Worst Movies

```python
from IPython.display import HTML # allows display in high quality website-like format
```

```python
# create seperate dataframe

df_best = df[["poster_path", "title", "budget_musd", "revenue_musd", "vote_count", "vote_average", "popularity"]].copy()
df_best
```

```python
# absolute profit (revenue - budget)
# roi (revenue / budget)
df_best["profit_musd"] = df.revenue_musd.sub(df.budget_musd)
df_best["return"] = df.revenue_musd.div(df.budget_musd)
df_best
```

```python
# renaming of labels
df_best.columns = ["", "Title", "Budget", "Revenue", "Votes", "Average Rating", "Popularity", "Profit", "ROI"]
```

```python
df_best
```

```python
# change the index to be the title column
df_best.set_index("Title", inplace=True)
```

```python
df_best
```

```python
df_best.iloc[0, 0] # disregards the index - we can get the images here themselves
```

```python
subset = df_best.iloc[:5, :2] # up to 5, up to 2
```

```python
subset
```

```python
HTML(subset.to_html(escape=False)) # no more html code oh wow
```

```python
df_best.sort_values(by="Average Rating", ascending=False)
```

```python
df_best.sort_values(by="ROI", ascending=False)
```

```python
# filter with budget greater and then sort
df_best.loc[df_best.Budget >= 5].sort_values(by = "ROI", ascending=False)
```

```python
df_best.Budget.fillna(0, inplace=True)
df_best.Votes.fillna(0, inplace=True)
```

```python
df_best.info()
```

```python
# best and worst n movies
def best_worst(n, by, ascending=False, min_bud=0, min_votes = 0):
    
    df2 = df_best.loc[(df_best.Budget >= min_bud) & (df_best.Votes >= min_votes), ["", by]].sort_values(by=by, ascending=ascending).head(n).copy()
    
    return HTML(df2.to_html(escape=False))
```

```python
best_worst(n=5, by="Revenue")
```

```python
best_worst(n=5, by="Budget")
```

```python
best_worst(n=5, by="Profit")
```

```python
best_worst(5, "Profit", ascending=True)
```

```python
best_worst(5, "ROI", min_bud=50)
```

```python
best_worst(5, "ROI", ascending=True, min_bud=20)
```

```python
best_worst(5, "Votes")
```

```python
best_worst(5, "Average Rating", min_votes=50)
```

```python
best_worst(5, "Average Rating", ascending=True, min_votes=50)
```

```python
best_worst(5, "Average Rating", ascending=True, min_votes=20, min_bud=20)
```

```python
best_worst(5, "Popularity")
```
